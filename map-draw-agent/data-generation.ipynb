```
# Extracting last two years of transactions
attrs = ['ci.transaction_id','ci.agent_uid','r.property_id', 'r.compass_property_id', 'listing_id', 'r.city', 'r.county', 'r.state','r.zipcode', 'r.list_date','r.close_date', 'r.retrieve_date', 'ci.user_id', 'r.county', 'r.latitude', 'r.longitude', 'ci.agent_name', 'r.list_price', 'r.close_price', 'r.close_date']

sql_ci = f'''
SELECT {" , ".join(attrs)}
    FROM silver_agent_data_v2.contact_info_unique_v2 as ci 
    join 
    (select * from listings_archive.root where retrieve_date = (SELECT MAX(retrieve_date) FROM listings_archive.root) ) r
    on ci.transaction_id = r.transaction_id
    WHERE coalesce(r.close_date,r.list_date)>DATE_ADD(now(),-1*365)
'''

print ("Fetching fresh data")
df = spark.sql(sql_ci).cache()
```
